```python
import csv
import pandas as pd
df = pd.read_csv("dataset.csv")
col=[]
for index, row in df.iterrows():
    label=""
    if row["sports"]==1 :
        label="sport"
    elif row["travel"]==1:
        label="travel"
    elif row["photography"]==1:
        label="photography"
    elif row["bola"]==1:
        label="bola"
    col.append(label)
#     print(label)
df["label"]=col
df.to_csv("dataset_update.csv")
```

```python
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split  
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix  

dataframe = pd.read_csv('dataset_update.csv')
dataframe =  dataframe.sample(frac=1).reset_index(drop=True)
X = dataframe['text'].values
y = dataframe['label'].values
tfidf_vectorizer = TfidfVectorizer(max_features=200000,use_idf=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)  
X_train = tfidf_vectorizer.fit_transform(X_train.astype('U'))
modelsvm = svm.LinearSVC()
modelsvm.fit(X_train,y_train)
x_test = tfidf_vectorizer.transform(X_test.astype('U'))
y_pred = modelsvm.predict(x_test)
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))  
print("accuracy ",modelsvm.score(x_test,y_test))
      
```
